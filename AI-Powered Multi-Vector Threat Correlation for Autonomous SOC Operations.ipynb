```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# 1. MULTI-VECTOR DATA (Simulated SOC Environment)
# Features: [Login_Hour, Data_Transfer_MB, Failed_Attempts, Typing_Deviation]
# A '1' in 'is_threat' means the combination of these is an attack.
data = {
    'hour':       [14, 15, 2, 14, 3, 16, 4],
    'data_mb':    [5, 10, 500, 8, 450, 12, 600],
    'fails':      [0, 0, 5, 0, 8, 1, 12],
    'type_dev':   [0.1, 0.2, 0.9, 0.1, 0.8, 0.2, 0.9],
    'is_threat':  [0, 0, 1, 0, 1, 0, 1] 
}
df = pd.DataFrame(data)

# 2. TRAIN THE SENTINEL BRAIN
X = df[['hour', 'data_mb', 'fails', 'type_dev']]
y = df['is_threat']
model = RandomForestClassifier(n_estimators=100, random_state=42).fit(X, y)

# 3. SOC CORRELATION FUNCTION
import pandas as pd
import warnings
warnings.filterwarnings("ignore") # This hides those annoying feature name warnings

def sentinel_scan(h, d, f, t, label):
    # We wrap the numbers in a small DataFrame so the names match
    test_df = pd.DataFrame([[h, d, f, t]], columns=['hour', 'data_mb', 'fails', 'type_dev'])
    
    prediction = model.predict(test_df)
    risk_prob = model.predict_proba(test_df)[0][1]
    
    print(f"--- SENTINEL CORRELATION REPORT: {label} ---")
    print(f"Combined Risk Score: {risk_prob * 100:.2f}%")
    
    if prediction[0] == 1:
        print("RESULT: [!!!] HIGH-LEVEL THREAT CORRELATED [!!!]")
        print("Action: Isolate Endpoint and Revoke Credentials.")
    else:
        print("RESULT: [✓] MULTI-VECTOR CHECK PASSED: SAFE")
    print("-" * 50)
# --- TEST THE BRAIN ---
sentinel_scan(14, 12, 0, 0.15, "Normal_Student_Activity") # Safe
sentinel_scan(3, 550, 10, 0.85, "Correlated_Attack_Pattern") # Dangerous
```
